**Ejercicio 1.-** Análisis y normalización de datos de sensores en un robot
móvilDataset a utilizar: robot_sensors.csv
Este dataset contiene mediciones de un robot móvil:
Tiempo (s): Marca de tiempo en segundos
Distancia (cm): Sensor de ultrasonido (distancia a un obstáculo)
Velocidad (cm/s): Velocidad del robot
Temperatura (°C): Temperatura del ambiente
Algunas filas contienen valores nulos o valores atípicos.

Implementar en una notebook de colab lo siguiente con este dataset:
Cargar el dataset con pandas

* Mostrar primeras filas del dataset
* Limpiar los datos
  * Eliminar filas con valores nulos
  * Filtrar valores atípicos (por ejemplo, si la velocidad es negativa)
* Normalizar los datos numéricos con Min-Max Scaling
* Visualizar los datos normalizados usando matplotlib



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url='https://raw.githubusercontent.com/dilp90/InteligenciaArtificial_y_RedesNeuronales_UANL_FIME/refs/heads/main/MachineLearning/Datasets/robot_sensors.csv'
```

```python
df= pd.read_csv('https://raw.githubusercontent.com/dilp90/InteligenciaArtificial_y_RedesNeuronales_UANL_FIME/refs/heads/main/MachineLearning/Datasets/robot_sensors.csv')
```

```python
print("Primeras filas del dataset:")
df = df.dropna() #se eliminan las filas que tengna datos nulos
clean = df[df['Velocidad (cm/s)'] > 10] #se eliminan velocidades anomalas (menos de 10cm/s)
print(clean.head(60)) #se imnprime la tabla
```

```python
clean.describe().loc[['min','max']]
```

```python
plt.figure()
plt.plot(clean["Tiempo (s)"], clean["Distancia (cm)"], label="Distancia")
plt.plot(clean["Tiempo (s)"], clean["Velocidad (cm/s)"], label="Velocidad")
plt.plot(clean["Tiempo (s)"], clean["Temperatura (°C)"], label="Temperatura")


plt.xlabel("Tiempo (s)")
plt.ylabel("Valores Normalizados")
plt.title("Datos de sensores")
plt.grid(True)
plt.legend()

plt.show()
```
